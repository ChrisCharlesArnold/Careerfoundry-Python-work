### Working with orders_products_merged

```python
#creating subset of the first million lines of data
df = df_ords_prods_mergd[:1000000]
```

```python
# creating a function to categorize items in to price ranges low, medium, high, and not enough data
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# creating a new column with our dataframe that will execute price_label on every row and output it to this column
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Checking categorization
df['price_range'].value_counts(dropna = False)
```

```python
#checking what the most expensive item is in our subset
df['prices'].max()
```

#### If-Statements with the loc() Function

```python
# performing the same idea as above but with the loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. ### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Checking head of dataframe to make sure we know proper column name to reference
df_ords_prods_mergd.head()
```

```python
#looking at frequency of orders for each hour of the day 
df_ords_prods_mergd['order_hour_of_day'].value_counts(dropna = False)
```

```python
# categorizing using loc (between hours of 9-17 inclusive)
df_ords_prods_mergd.loc[(df_ords_prods_mergd['order_hour_of_day'] <= 17 ) & (df_ords_prods_mergd['order_hour_of_day'] >9 ) , 'busiest_period_of_day'] = 'Most Orders'
```

```python
# categorizing using loc (greater than hour 17 )
df_ords_prods_mergd.loc[((df_ords_prods_mergd['order_hour_of_day'] > 17 )), 'busiest_period_of_day'] = 'Average Orders'
```

```python
# categorizing using loc (between the hours of 6 and 9)
df_ords_prods_mergd.loc[((df_ords_prods_mergd['order_hour_of_day'] <= 9 ) & (df_ords_prods_mergd['order_hour_of_day'] >= 6)), 'busiest_period_of_day'] = 'Average Orders'
```

```python
# categorizing using loc (greater than or equal to the hour of 0 and less than 2 not inclusive)
df_ords_prods_mergd.loc[((df_ords_prods_mergd['order_hour_of_day'] >= 0 ) & (df_ords_prods_mergd['order_hour_of_day'] < 2)), 'busiest_period_of_day'] = 'Average Orders'
```

```python
# categorizing using loc (greater than 1 (not inclusive) and less than or equal to 5 (inclusive))
df_ords_prods_mergd.loc[((df_ords_prods_mergd['order_hour_of_day'] > 1 ) & (df_ords_prods_mergd['order_hour_of_day'] <= 5 )), 'busiest_period_of_day'] = 'Fewest orders'
```

### Print the frequency for this new column

```python
#checking categorization counts
df_ords_prods_mergd['busiest_period_of_day'].value_counts(dropna = False)
```
